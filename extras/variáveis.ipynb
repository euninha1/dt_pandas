```python
import pandas as pd 

dados = pd.read_csv('dados/aluguel.csv', sep = ';')
```

```python
dados
```

```python
#criando variável do valor bruto 

dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados
```

```python
#a função round é usada para arrendodar numeros com casas decimais 


dados['Valor m2'] = dados['Valor m2'].round(2)
dados
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados
```

```python
casa = ['Casa','Casa de condomínio', 'Casa de Vila']
```

```python
#o método apply() permite que apliquemos uma função à cada registro do DataFrame.

dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

```python
import pandas as pd
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas'])
```

```python
alunos['Faixa Etária'] = alunos['Idade'] \
    .apply(lambda x: 'Menor que 20 anos' if x < 20 
        else ('Entre 20 e 40 anos' if (x >= 20 and x <= 40) 
            else 'Maior que 40 anos'))
```

```python
alunos['Notas-Média(Notas)'] = alunos['Notas'] \
    .apply(lambda x: x - alunos['Notas'].mean())
```

```python
alunos['Notas-Média(Notas)'] = \
    alunos.Notas - alunos.Notas.mean()
```
